## ***`IMPORTING LIBRARIES`***

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

# ***`LOADING DATASET`***

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/income_evaluation.csv')
```

```python
df.shape
df.head()
```

```python
df.isnull().sum()
```

# ***RANDOM FOREST MODEL***

```python
x=df[['age',' workclass',' fnlwgt',' sex',' hours-per-week',' occupation']]
y=df[' income']

x.loc[:,' sex']=x[' sex'].map({'Male':1,'Female':0})

# Correct column names by stripping whitespace
x.columns = x.columns.str.strip()

# Apply one-hot encoding to categorical features
x = pd.get_dummies(x, columns=['workclass', 'occupation'], drop_first=True)


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

rf_classifier=RandomForestClassifier(n_estimators=100,random_state=42)
rf_classifier.fit(x_train,y_train)

y_pred=rf_classifier.predict(x_test)

accuracy=accuracy_score(y_test,y_pred)
classification_rep=classification_report(y_test,y_pred)
print(f"Accuracy:{accuracy:.2f}")
print(f"Classification Report:\n{classification_rep}")

sample=x_test.iloc[0:1]
predicition=rf_classifier.predict(sample)

sample_dict=sample.iloc[0].to_dict()
print(f"\n Sample person:{sample_dict}")
print(f"Predicted Income:{'>50k'if predicition[0]==1 else '<=50k'}")
```
